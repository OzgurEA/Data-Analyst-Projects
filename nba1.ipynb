```python
top_players = nba1.nlargest(10, ["pts", "ast", "reb"])

top_players["total_score"] = top_players["pts"] + top_players["ast"] + top_players["reb"]

top_players = top_players.sort_values("total_score", ascending=False)

print(top_players[["player_name", "pts", "ast", "reb", "season", "Total"]])
```

```python
stats = ['pts', 'reb', 'ast']
weights = [1, 0.7, 0.8]

nba1['total_score'] = nba1[stats].mul(weights).sum(axis=1)
nba1_sorted = nba1.sort_values(by=['total_score', 'pts', 'ast', 'reb'], ascending=False)

print(nba1_sorted[['player_name', 'total_score', 'pts', 'ast', 'reb', 'season']].head(10))
```

```python
stats = ["pts", "ast", "reb"]
weights = [1, 1, 1]

nba1["total_score"] = nba1[stats].mul(weights).sum(axis=1)
nba1_sorted = nba1.sort_values(by=["total_score", "pts", "ast", "reb"], ascending=False)

print(nba1_sorted[["player_name", "total_score", "pts", "ast", "reb", "season"]].head(10))
```

The code we wrote here is a bit complicated and has several layers. We wanted to rank players according to their "pts", "ast" and "reb" stats. The peculiarity of the ranking we did was to select the highest player in each area. The questions of who scored the most points, assists and rebounds are nice, but a bit simplistic. But isn't there someone who excels in all of them? There certainly are. As you can see, Russell Westbrook dominated the 2017-18 season, averaging a triple-double every game for an entire season. But what about the others? As you can see, there are two codes that do the same thing but with different results.  So what is this difference and what does it tell us? First of all, what it tells us is that data manipulation will not always give us precise results and that there is a gray area. We can shape it however we want. As for the difference, the main reason for this is prioritization. Code cell 8 has the priority "pts". But in code cell 9 things change a little bit. With the "weights" assignment, we tied the priority to a coefficient and determined the "total_score" result according to the "weights" assignment with the stats. That's the main difference. However, even though there is no difference between the coefficient values, the ordering is different between these codes (for example, I wrote another code where the values of the "weights" assignment are 1). This supports the prioritization I mentioned. So, if you ask which code is better to use, I can't say anything clear. Both of these codes make the most correct ordering in their own way, but the second code seems more correct to me because we have much more authority over the code.

```python
player_name = "James Harden"

player_data = nba1[nba1["player_name"] == player_name]
player_data = player_data.sort_values("season")

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(player_data["season"], player_data["pts"], label="Points", marker="o")
ax.plot(player_data["season"], player_data["ast"], label="Asist", marker="o")
ax.plot(player_data["season"], player_data["reb"], label="Rebounds", marker="o")

ax.set_xlabel("Season")
ax.set_ylabel("Stats")
ax.set_title(f'{player_name} Season Performance')
ax.legend(loc="best")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Harden has an interesting statistic as seen in our previous tables, so I wanted to analyze it. What is this guy's success and when did it start? And so we started writing code that started to get a bit more complex. And here is the graph of Harden's stats. What we read from this graph is that Harden was a scorer from the very beginning. His rebounding and assists haven't changed much, but they have improved over the years. But as you can see, he has made an incredible leap in average scoring and has taken his place at the top. So where did this anomaly come from? There are several ways to investigate this, but we need to add the teams the players played for to this dataset so that we can make a comparison with their teammates. Unfortunately, this dataset does not include that data. Therefore, I will do this comparison research with Stephen Curry, one of Harden's biggest rivals and my favorite player:

```python
player_name1 = "James Harden"
player_name2 = "Stephen Curry"

player_data1 = nba1[nba1["player_name"] == player_name1]
player_data2 = nba1[nba1["player_name"] == player_name2]

player_data1 = player_data1.sort_values("season")
player_data2 = player_data2.sort_values("season")

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 18), sharex=True)
ax1.plot(player_data["season"], player_data1["pts"], label=player_name1, marker="o")
ax1.plot(player_data["season"], player_data2["pts"], label=player_name2, marker="o")
ax1.set_ylabel("Points")
ax1.legend(loc="best")

ax2.plot(player_data["season"], player_data1["ast"], label=player_name1, marker="o")
ax2.plot(player_data["season"], player_data2["ast"], label=player_name2, marker="o")
ax2.set_ylabel("Asist")
ax2.legend(loc="best")

ax3.plot(player_data["season"], player_data1["reb"], label=player_name1, marker="o")
ax3.plot(player_data["season"], player_data2["reb"], label=player_name2, marker="o")
ax3.set_xlabel("Season")
ax3.set_ylabel("Rebound")
ax3.legend(loc="best")

fig.suptitle(f'{player_name1} vs {player_name2} Season Performance Comparison', fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

Yes, the statistics seen in these comparative graphs are real, even though they make me sad. First of all, Harden has shown a more steady improvement in assists and rebounds than Curry. While Curry's Assists and Rebounds stats are less than Harden's, they have not changed much. Of course, the 15-16 season break in the assist chart should also be taken into consideration. But I can't say the same for the scoring situation, where Harden has been above Curry for a long time, even though they have been scoring close to each other. Even though Curry finished above Harden in the 15-16 season, in the following seasons Curry had a serious decline and Harden had a serious rise. Until the 19-20 season. After that season, Curry took the initiative and surpassed Harden. So what does this data tell us in the first place? That championships can be won with a team.
